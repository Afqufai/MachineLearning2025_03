# Tugas - Wisconsin Breast Cancer
Pada tugas pratikum ini Anda akan menggunakan data "Wisconsin Breast Cancer". Data tersebut terdiri dari 569 data yang digunakan untuk mendiagnonis jenis kanker Malignant (M) dan Benign (B). Tugas Anda adalah,

- Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

- Lakukan proses encoding pada kolom "diagnosis".

- Lakukan proses standardisasi pada semua kolom yang memiliki nilai numerik.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/Dataset/wbc.csv')
df.head()
```

Slice/Pisahkan Data yang tidak dapat digunakan

```python
# Slice data yang tidak relevan dan tidak dapat digunakan
df = df.drop(['id', 'Unnamed: 32'], axis=1)
df.head()
```

Encode kolom Diagnosies

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # buat dari LabelEncoder
df['diagnosis'] = le.fit_transform(df['diagnosis']) # proses encoding
df.head()
```

Lakukan standarisasi semua yang punya nilai numerik

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
# Standarisasi SEMUA dengan nilai NUMERIK
numeric_cols = df.select_dtypes(include=np.number).columns.tolist() # Select kolom numerik
numeric_cols.remove('diagnosis') # Hapus kolom diagnosis
df[numeric_cols] = std.fit_transform(df[numeric_cols]) # Standarisasi
df.head(20)
```
